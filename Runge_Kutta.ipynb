
# Runge-Kutta4

```python
def RungeKutta4(x, c, step_size,list):
    ''' método de Runge-Kutta de cuarto orden para resolver ecuaciones diferenciales
        calcula el valor de la función en cuatro puntos diferentes y los gradientes correspondientes.
        Finalmente, actualiza el valor de x utilizando la fórmula de Runge-Kutta de cuarto orden.
    '''

    x1 = x.detach().clone().requires_grad_(True)
    y1 = c(x1)
    list.append(y1.item())
    y1.backward()
    k1 = -1*x1.grad.clone()
    x1.grad.data.zero_()

    x2 = (x1 + k1 * step_size / 2).detach().clone().requires_grad_(True)
    y2 = c(x2)
    y2.backward()
    k2 = -1*x2.grad.clone()
    x2.grad.data.zero_()

    x3 = (x1 + k2 * step_size / 2).detach().clone().requires_grad_(True)
    y3 = c(x3)
    y3.backward()
    k3 = -1*x3.grad.clone()
    x3.grad.data.zero_()

    x4 = (x1 + k3 * step_size).detach().clone().requires_grad_(True)
    y4 = c(x4)
    y4.backward()
    k4 = -1*x4.grad.clone()
    x4.grad.data.zero_()


    with torch.no_grad():
      x+=  ((k1 + 2*k2 + 2*k3 + k4) * step_size / 6)
```
