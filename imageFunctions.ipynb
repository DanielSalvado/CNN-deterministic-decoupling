```python
from torchvision import transforms
import matplotlib.pyplot as plt
import torch
```

```python
def prepare_image(image,size=0):
  '''Ajusta el tamaÃ±o de una textura y la transforma en un tensor'''
  if size > 0:
    image=image.crop((0, 0, size, size))
  image = transforms.ToTensor()(image)
  image = image.to(torch.float32)

  return image.unsqueeze(0)
```

```python
def show_image(image):
  '''Transforma un tensor en imagen y la muestra por pantalla'''
  image = image.squeeze(0)
  image = transforms.ToPILImage()(image)
  plt.imshow(image)
  plt.axis('off')
  plt.show()
```

```python
def save_image(image, name):
  '''Transforma un tensor en imagen y la guarda en un directorio'''
  image = image.to(torch.float32)
  image = image.squeeze(0)
  image= transforms.ToPILImage()(image)
  image.save(name)
```
